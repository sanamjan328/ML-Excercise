## 2. Exploratory Data Analaysis

```python
plt.figure(figsize=(16,8))

plt.subplot(233)
sns.countplot(y='disease_type',data=df)
plt.gca().xaxis.tick_bottom()
plt.title('Data count by Class')
```

```python
print("Any missing sample in data set:",df.isnull().values.any())
```

```python
from sklearn import preprocessing
X = pd.DataFrame(df.drop(['Sample','disease_type'], axis=1))
y = df.disease_type.values.astype(object)
encoder = preprocessing.LabelEncoder()
encoder.fit(y)
y = encoder.transform(y)
print('The independent features set: ')
print(X)
print('The dependent variable: ')
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.externals import joblib
import scikitplot as skplt
```

```python
## split the training data and testing data with 75% train 25% test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=21)
```

```python
## preprocessing Feature Scaling 
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

## 3. Prediction and Evaluation

### 3.1 Training Model_1

```python
rfc = RandomForestClassifier(n_estimators = 600, criterion = 'entropy', random_state = 42)
rfc.fit(X_train_scaler, y_train)
```

### 3.2 Create Random Forest Classification report from the results

```python
rfc_predict = rfc.predict(X_test_scaler)
cr_rfc = classification_report(y_test, rfc_predict)
print(cr_rfc)
```

### 3.3. Show the Random Forest Classification Confiusion Matrix for the prediction

```python
rfc_cm = confusion_matrix(y_test, rfc_predict)
print(rfc_cm)
```

```python
skplt.metrics.plot_confusion_matrix(y_test, rfc_predict, figsize=(10, 8))
plt.show()
```

### 3.4 Training Model_2

```python
dtree = DecisionTreeClassifier(criterion='gini', max_depth=None)
dtree.fit(X_train_scaler, y_train)
```

### 3.5 Create Decision Tree Classification report from the result

```python
dtree_predict = dtree.predict(X_test_scaler)
cr_dtree = classification_report(y_test, dtree_predict)
print(cr_dtree)
```

```python
dtree_cm = confusion_matrix(y_test, dtree_predict)
print(dtree_cm)
```

```python
skplt.metrics.plot_confusion_matrix(y_test, dtree_predict, figsize=(10, 8))
plt.show()
```

### 3.8 Training Model_4

```python
from sklearn.utils import shuffle
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
print("Dimension of Train set",X_train.shape)
print("Dimension of Test set",X_test.shape,"\n")
```

```python
params_grid = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4],
                     'C': [1, 10, 100, 1000]},
                    {'kernel': ['linear'], 'C': [1, 10, 100, 1000]}]
```

```python
## parameters using grid search and cross validation
svm_model = GridSearchCV(SVC(), params_grid, cv=5)
svm_model.fit(X_train_scaler, y_train)
```

### 3.9 Create Suport Vectore Machine Classification report from the results

```python
print('Best score for training data:', svm_model.best_score_,"\n") 
print('Best C:',svm_model.best_estimator_.C,"\n") 
print('Best Kernel:',svm_model.best_estimator_.kernel,"\n")
print('Best Gamma:',svm_model.best_estimator_.gamma,"\n")

final_model = svm_model.best_estimator_
svm_predict = final_model.predict(X_test_scaler)
cr_svm = classification_report(y_test, svm_predict)
print(cr_svm)
```

```python
svm_cm = confusion_matrix(y_test, svm_predict)
print(svm_cm)
```
