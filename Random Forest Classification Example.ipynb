## Set up Data

### Categorical Feature
The purpose column as catagorical, we transform the using dummy variable to sklearn will be to understand them

```python
df_final = pd.get_dummies(df, ['purpose'], drop_first=True)
```

```python
df_final.head(8)
```

### Split Train and Test data

```python
from sklearn.model_selection import train_test_split
X = df_final.drop('not.fully.paid', axis=1)
y = df_final['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
X.head()
```

## Training a Deision Tree Model
### Create an instance of DecisionTreeClassifler() called dtree and fit it to the training data set

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(criterion='gini', max_depth=None)
dtree.fit(X_train, y_train)
```

## Prediction and Evalution of Decision Tree
### create prediction from the test set and create a classification report and a confusion matrix

```python
predictions = dtree.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, predictions))
```

```python
cm = confusion_matrix(y_test, predictions)
print(cm)
print('Accurancy of prediction:', round((cm[0,0]+cm[1,1])/cm.sum(), 3))
```

## Training the Random Forest model
### Create Instance of the RandomForestClassifier class and fit it to our training data from the previous step

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=600)
rfc.fit(X_train, y_train)
```

## Predictions and Evaluation
### create a classification report from the results, do you get anything strange or sort of warning

```python
rfc_predict = rfc.predict(X_test)
cr = classification_report(y_test, predictions)
print(cr)
```

### Show the Confiusion Matrix for the prediction 

```python
cm = confusion_matrix(y_test, rfc_predict)
print(cm)
```

## Running a loop with increasing number of trees in the random forest and checking accuracy of confusion matrix

### Criterion 'gini' or 'entropy'

```python
nsimu = 21
accuracy=[0]*nsimu
ntree = [0]*nsimu
for i in range(1, nsimu):
    rfc = RandomForestClassifier(n_estimators=i*5, min_samples_split=10, max_depth=None, criterion='gini')
    rfc.fit(X_train, y_train)
    rfc_predict = rfc.predict(X_test)
    cm = confusion_matrix(y_test, rfc_predict)
    accuracy[i] = (cm[0,0]+cm[1,1])/cm.sum()
    ntree[i]=i*5
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=ntree[1:nsimu], y=accuracy[1:nsimu], s=60, c='red')
plt.title("Number of trees in the Random Forest vs. prediction accuracy (criterion: 'gini')", fontsize=18)
plt.xlabel("Number of trees", fontsize=15)
plt.ylabel("Prediction accuracy from confusion matrix", fontsize=15)
```
